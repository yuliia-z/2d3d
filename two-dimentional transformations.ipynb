```python
from matplotlib import pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
```

```python
points = [
    (1., 1.),
    (1., 2.),
    (3., 2.),
    (3., 1.),
    (1., 1.),
    (1., 2.),
    (2., 3.),
    (3., 2.)
    ]

paths = [Path.MOVETO,
         Path.LINETO,
         Path.LINETO,
         Path.LINETO,
         Path.CLOSEPOLY,
         Path.MOVETO,
         Path.LINETO,
         Path.LINETO,
         ]
```

```python
path = Path(points, paths)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
patch = patches.PathPatch(path, facecolor=(0,0,0,0), lw=2)
ax.add_patch(patch)
ax.set_xlim(0,4)
ax.set_ylim(0,4)

plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
