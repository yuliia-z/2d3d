```python
from matplotlib import pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
import numpy as np
```

```python
global points

points = [
    (1., 1.),
    (3., 1.),
    (1., 5.),
    (1., 1.)
    ]

paths = [
         Path.MOVETO,
         Path.LINETO,
         Path.LINETO,
         Path.LINETO]
```

```python
def scale(xscale, yscale):
    global points
    return [(point[0] * xscale, point[1] * yscale) for point in points]
```

```python
scaled_points = scale(3, 3)

patch = patches.PathPatch(Path(points, paths), facecolor='orange', lw=2)
scaled_patch = patches.PathPatch(Path(scaled_points, paths), facecolor='blue', lw=2)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

ax.add_patch(patch)
ax.add_patch(scaled_patch)

ax.set_xlim(0,15)
ax.set_ylim(0,15)

plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
